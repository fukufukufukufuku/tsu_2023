# リストの書き方
- `リスト`は、複数の値をまとめて扱う場合に使用
- `リスト`は、数値や文字列などの基本データ型と並んでPythonでは高頻度で使われるデータ型
- `リスト`は全体を`[]`で囲み、各要素は`カンマ`で区切る
- `リスト`は型の値でも要素として格納することができ、リストの中にリストを格納することも可能である 
- `リスト`では`インデックス`(要素番号)を用いて、要素の追加や入れ替えなどが可能

## 要素の削除

### インデックスによる要素の削除
```Python
リスト.pop(削除する要素のインデックス)
```

````python
# 削除前のリストを表示
listA = [90, 85, 75, 95, 80] # 5人の点数
print("削除前:", listA)

# インデックス(要素番号)0番目を削除
listA.pop(0)

# 削除後のリストを表示
print("削除後:", listA)
````

### 特定の値の要素の削除
```
リスト.remove(削除する要素)`
```
- リストの中に該当する要素が見つからなかった場合はエラー(ValueError)が発生する
- 削除される要素は「最初に該当したもの」のため、同じ値が2回以上出てくると最初に出る要素だけ削除される

````python
# 削除前のリストを表示
listC = [90, 85, 75, 95, 85] # 5人の点数
print("削除前:", listC)

# 1つ目の85点を削除
listC.remove(85)

# 削除後のリストを表示
print("1回目削除後:", listC)

# 2つ目の85点を削除
listC.remove(85)

# 削除後のリストを表示
print("2回目削除後:", listC)
````

# リストのサイズの操作

## 複数のリストの結合
`リストA = リストB + リストC`

**最初のテストの受験者5人の結果は、90点、85点、75点、95点、80点であったとする。その後、新たに4人がテストを受験して、70点, 80点, 75点, 90点であった。さらに、2人がテストを受験して85点、95点であった。**

````python
listA = [90, 85, 75, 95, 80]
listB = [70, 80, 75, 90]
listC = [85, 95]

# listAとlistBを結合させてlistDを作成
listD = listA + listB

# リストを出力
print("1回目の結合：", listD)

# listDに対してlistCを結合させてlistDを拡張
listD = listD + listC     # listD += listC でも可

# リストを出力
print("2回目の結合：", listD)
````

# リストの要素の並び替え
- `sorted関数`を用いる
  - 昇順に並び替え（小→大）
  ```Python
  sorted(リスト)
  ```
  - 降順に並び替え（大→小）
  ```Python  
  sorted(リスト,reverse=True)` 
  ```    

## 要素が数値の場合

````python
listD = [90, 85, 75, 95, 80, 70, 80, 75, 90, 85]

# listDを昇順に並び替えてlistEに代入
listE = sorted(listD)
print("昇順に並び替え: ", listE)

# listDを降順に並び替えてlistFに代入
listF = sorted(listD, reverse=True)
print("降順に並び替え: ", listF)
````

## 要素が文字列の場合(アルファベット順)

````python
listG = ["dog", "cat", "tiger", "bird", "horse"]

listH = sorted(listG)
print("昇順に並び替え: ", listH)

listI = sorted(listG, reverse=True)
print("降順に並び替え: ", listI)
````

# for文によるリストの要素の操作
```Python
for 要素を代入する変数 in リスト:
    実行する処理
```
要素を1個ずつ変数に代入して繰り返す

**10人のテストの点数の合計値と平均値を計算**<br>

````python
 # 10人のテストの点数を格納
listD = [90, 85, 75, 95, 80, 70, 80, 75, 90, 85]

total = 0               # 合計値の初期値を0とする
for i in listD:        # 要素を１個ずつ取り出して変数eleに代入
    total = total + i  # 要素を足す　

# 平均値 = 合計/要素数    
average = total / len(listD)  # 平均値をaverageという変数に代入

print("合計値", total)
print("平均値", average)
````

# リスト内の要素の統計計算
- 標準ライブラリの`statistics`を用いて、平均、中央値、最頻値、分散、標準偏差を算出できる
  - ライブラリとは、よく使う機能・関数をまとめて、簡単に使えるようにしたもの
  - ライブラリの中身のプログラムがブラックボックスであっても、入力と出力がわかっていれば簡単に実装し、実行することができる
  - Pythonに便利な機能を持つライブラリが多数ある
- 標準ライブラリを使う場合、import文で呼び出す
```Python
import statistics
statistics.mean(リスト)      # 平均値
statistics.median(リスト)    # 中央値(要素を小さい方から大きい方に並べたときの中央の値)
statistics.median(リスト)    # 最頻値(最も出現頻度が高い値)
statistics.pvariance(リスト) # 分散(データのばらつき)
statistics.pstdev(リスト)    # 標準偏差(分散の平方根)
```

**10人のテストの点数の統計量を計算**

````python
import statistics

# 10人の点数
listJ = [90, 85, 75, 95, 80, 70, 75, 90, 80, 75] 

mean = statistics.mean(listJ)
print("平均値: ",  mean)

median = statistics.median(listJ)
print("中央値: ", median)

mode = statistics.mode(listJ)
print("最頻値: ", mode)

pvariance = statistics.pvariance(listJ)
print("分散: ", pvariance) 

pstdev = statistics.pstdev(listJ)
print("標準偏差: ", pstdev)
````
